```python
#Import packages
import pandas as pd
import re, string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer 
from textblob import TextBlob
sw = stopwords.words('english')
```

```python
#The function
def clean_text(text):
    text = text.lower()
    text = re.sub('@', '', text)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub(r"[^a-zA-Z ]+", "", text)
    
    #Tokenize the data
    text = nltk.word_tokenize(text)
    #Remove stopwords
    text = [w for w in text if w not in sw]
    return text
```

```python
extracted = comments[['title','content'].copy()]
extracted.head()
```

```python
extracted['title'] = extracted['title'].apply(lambda x: clean_text(x)) 
extracted['content'] = extracted['content'].apply(lambda x: clean_text(x))    
```

```python
extracted['title']
```

```python
#Lemmatizer
lemmatizer = WordNetLemmatizer()
def lem(text):
    text = [lemmatizer.lemmatize(t) for t in text]
    text = [lemmatizer.lemmatize(t, 'v') for t in text]
    return text
```

```python
nltk.download('wordnet')
extracted['content'] = extracted['content'].apply(lambda x: lem(x))
extracted['title'] = extracted['title'].apply(lambda x: lem(x))
```

```python
extracted['content']
```

```python
#From lists of comments to a single list containing all words      
all_words=[]        
for i in range(len(comments)):
    all_words = all_words + extracted['title'][i]
#Get word frequency        
nlp_words = nltk.FreqDist(all_words)
plot1 = nlp_words.plot(20, color='salmon', title='Word Frequency')
```

```python
#Bigrams
bigrm = list(nltk.bigrams(all_words))
words_2 = nltk.FreqDist(bigrm)
words_2.plot(20, color='salmon', title='Bigram Frequeancy')
```

```python
#Trigram
trigrm = list(nltk.trigrams(all_words))
words_2 = nltk.FreqDist(trigrm)
words_2.plot(20, color='salmon', title='Trigram Frequeancy')
```

```python
#Get sentiment from comments
comments['analysis'] = [str(thing) for thing in comments['analysis']]
sentiment = []
for i in range(len(comments)):
    blob = TextBlob(comments['analysis'][i])
    for sentence in blob.sentences:
        sentiment.append(sentence.sentiment.polarity)
comments['analysis']=sentiment
#Plot
comments['analysis'].plot.hist(color='salmon', title='Comments Polarity')
```
